```python
import sys
sys.path.append('..')
import numpy as np
import metatensor as mt
import matplotlib.pyplot as plt
import matplotlib
import time
```

```python
"""
Generate training data.
"""
pic = matplotlib.image.imread('../data/Mondrian/mondrian.jpg') / 255
w, h = pic.shape

"""
Hyper-parameters.
"""
num_epoch = 50
lr = 0.01 # learning rate
batch_size = 10
```

# Sobel Filter

```python
# sobel filter
sobel_v = mt.core.Variable(dim=(3, 3), init=False, trainable=False) # vertical
sobel_h = mt.core.Variable(dim=(3, 3), init=False, trainable=False) # horizontal

sobel_v.set_value(np.mat([[1, 0, -1], [2, 0, -2], [1, 0, -1]]))
sobel_h.set_value(sobel_v.value.T)

# input
img = mt.core.Variable(dim=(w, h), init=False, trainable=False)
img.set_value(np.mat(pic))

# sobel output
sobel_v_output = mt.ops.Convolve(img, sobel_v)
sobel_h_output = mt.ops.Convolve(img, sobel_h)

square_output = mt.ops.Add(
    mt.ops.Multiply(sobel_v_output, sobel_v_output),
    mt.ops.Multiply(sobel_h_output, sobel_h_output)
)

square_output.forward()

# output
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(221)
ax.axis('off')
ax.imshow(square_output.value, cmap='gray')

ax = fig.add_subplot(223)
ax.axis('off')
ax.imshow(sobel_v_output.value, cmap='gray')

ax = fig.add_subplot(223)
ax.axis('off')
ax.imshow(sobel_h_output.value, cmap='gray')
```

# Train Sobel Filter

```python
sobel_v = mt.core.Variable(dim=(3, 3), init=False, trainable=False) # vertical
sobel_v.set_value(np.mat([[1, 0, -1], [2, 0, -2], [1, 0, -1]]))

img = mt.core.Variable(dim=(w, h), init=False, trainable=False)
img.set_value(np.mat(pic))

sobel_output = mt.ops.Convolve(img, sobel_v)
sobel_output.forward()
plt.imshow(sobel_output.value, cmap='gray')
```

```python
filter_trainable = mt.core.Variable(dim=(3, 3), init=True, trainable=True)
filter_output = mt.ops.Convolve(img, filter_trainable)

minus = mt.core.Variable(dim =(w, h), init=False, trainable=False)
minus.set_value(np.mat(-np.ones((w, h))))

# error = sobel_output - filter_output
error = mt.ops.Add(sobel_output, mt.ops.Multiply(filter_output, minus))
square_error = mt.ops.MatMul(mt.ops.Reshape(error, shape=(1, w * h)), mt.ops.Reshape(error, shape=(w * h, 1)))

n = mt.core.Variable(dim=(1, 1), init=False, trainable=False)
n.set_value(np.mat(1.0 / (w * h)))
mse = mt.ops.MatMul(square_error, n)

optimizer = mt.optimizer.Adam(mt.core.default_graph, mse, lr)

for i in range(num_epoch):
    start_time = time.time()
    optimizer.one_step()
    optimizer.update()
    mse.forward()
    print('iteration:{:d}, loss:{:.6f}, time cost:{:.3f}'.format(i, mse.value[0, 0], time.time()-start_time))
```
